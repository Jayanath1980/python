```python
class Student:
    def __init__(self, student_id, name, grade):
        self.student_id = student_id
        self.name = name
        self.grade = grade

    def set_student_id(self, student_id):
        self.student_id = student_id

    def get_student_id(self):
        return self.student_id

    def set_name(self, name):
        self.name = name

    def get_name(self):
        return self.name

    def set_grade(self, grade):
        self.grade = grade

    def get_grade(self):
        return self.grade

# Test the Student class
student_id = input("Enter student ID: ")
name = input("Enter student name: ")
grade = input("Enter student grade: ")

student = Student(student_id, name, grade)

# Get and print student attributes
print("Student ID:", student.get_student_id())
print("Student Name:", student.get_name())
print("Student Grade:", student.get_grade())

# Modify student attributes
new_id = input("Enter new student ID: ")
student.set_student_id(new_id)

new_name = input("Enter new student name: ")
student.set_name(new_name)

new_grade = input("Enter new student grade: ")
student.set_grade(new_grade)

# Get and print modified student attributes
print("Modified Student ID:", student.get_student_id())
print("Modified Student Name:", student.get_name())
print("Modified Student Grade:", student.get_grade())
```

```python
class BankAccount:
    def __init__(self, account_number, initial_balance=0):
        self.account_number = account_number
        self.balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount}. New balance: ${self.balance}")
        else:
            print("Invalid deposit amount. Amount must be greater than 0.")

    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.balance:
                self.balance -= amount
                print(f"Withdrew ${amount}. New balance: ${self.balance}")
            else:
                print("Insufficient funds. Withdrawal amount exceeds balance.")
        else:
            print("Invalid withdrawal amount. Amount must be greater than 0.")

    def display_balance(self):
        print(f"Account Number: {self.account_number}, Balance: ${self.balance}")


# Test the BankAccount class
account_number = input("Enter account number: ")
initial_balance = float(input("Enter initial balance: "))

# Create BankAccount instance
account = BankAccount(account_number, initial_balance)

# Deposit some money
amount_to_deposit = float(input("Enter amount to deposit: "))
account.deposit(amount_to_deposit)

# Withdraw some money
amount_to_withdraw = float(input("Enter amount to withdraw: "))
account.withdraw(amount_to_withdraw)

# Display account balance
account.display_balance()

```

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def __str__(self):
        return f"{self.title} by {self.author}"


class User:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return self.name


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print(f"Book '{book.title}' by {book.author} added to the library.")

    def remove_book(self, book):
        if book in self.books:
            self.books.remove(book)
            print(f"Book '{book.title}' by {book.author} removed from the library.")
        else:
            print("Book not found in the library.")

    def lend_book(self, book, user):
        if book in self.books:
            self.books.remove(book)
            print(f"Book '{book.title}' by {book.author} lent to {user}.")
        else:
            print("Book not available for lending.")


# Test the Library Management System
library = Library()

# Add some books to the library
book1 = Book("The Great Gatsby", "F. Scott Fitzgerald")
book2 = Book("To Kill a Mockingbird", "Harper Lee")
library.add_book(book1)
library.add_book(book2)

# Create a user
user = User("Alice")

# Lend a book to the user
library.lend_book(book1, user)

# Remove a book from the library
library.remove_book(book2)

# Try to lend a book that is not in the library
library.lend_book(book2, user)
```

```python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, name, quantity, price):
        item = {"name": name, "quantity": quantity, "price": price}
        self.items.append(item)
        print(f"Added {quantity} {name}(s) to the shopping cart.")

    def remove_item(self, name, quantity):
        for item in self.items:
            if item["name"] == name:
                if item["quantity"] >= quantity:
                    item["quantity"] -= quantity
                    print(f"Removed {quantity} {name}(s) from the shopping cart.")
                    if item["quantity"] == 0:
                        self.items.remove(item)
                    return
                else:
                    print(f"Could not remove {quantity} {name}(s). Not enough in the cart.")
                    return
        print(f"{name} is not in the shopping cart.")

    def calculate_total_price(self):
        total_price = sum(item["quantity"] * item["price"] for item in self.items)
        return total_price


# Test the ShoppingCart class
cart = ShoppingCart()

# Add items to the shopping cart
cart.add_item("Apples", 3, 1.50)
cart.add_item("Bananas", 2, 0.75)
cart.add_item("Oranges", 4, 1.25)

# Calculate and print the total price
total_price = cart.calculate_total_price()
print(f"Total price of items in the cart: ${total_price:.2f}")

# Remove some items from the shopping cart
cart.remove_item("Apples", 2)
cart.remove_item("Bananas", 3)

# Calculate and print the updated total price
total_price = cart.calculate_total_price()
print(f"Updated total price of items in the cart: ${total_price:.2f}")
```
